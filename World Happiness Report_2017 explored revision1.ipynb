```python
#plt.rcParams['figure.figsize'] = (10,2)
#health = df.HealthLifeExpectancy.head(20).plot(kind='hist')
health = df.HealthLifeExpectancy.head(20)
#binomial_two_sample1000 = binomial_two[0:1000] 

plt.hist(health)


# Add a vertical line at the mean.
#plt.axvline(rand1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(health.mean(), color='r', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
#plt.axvline(rand1.mean() + rand1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(health.mean() + health.std(), color='r', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
#plt.axvline(rand1.mean()-rand1.std(), color='b', linestyle='dashed', linewidth=2) 
plt.axvline(health.mean() - health.std(), color='r', linestyle='dashed', linewidth=2)


plt.ylabel('Number of Countries')
plt.xlabel('Age')
plt.title('20 Happiest Counrties: Life Expectancy')
plt.show()
print('Mean:')
print(health.mean())
print('Standard Deviation:')
print(health.std())
```
